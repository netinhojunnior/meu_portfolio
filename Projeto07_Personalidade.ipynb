```python
# Carreganndo bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
# Carregando base de dados

dados = pd.read_csv('/content/drive/MyDrive/Portfólio/Dados/personality_dataset.csv')
```

```python
# Visualizando dados

dados.head()
```

**Informações sobre as Variáveis**

- **Time_spent_Alone:** Horas gastas sozinho diariamente (0–11).
- **Stage_fear:** Presença de medo do palco (Yes/No).
- **Social_event_attendance:** Frequency of social events (0–10).
- **Going_outside:** Frequência de saídas de casa (0–7).
- **Drained_after_socializing:** Sentimento de esgotamento após socializar (Yes/No).
- **Friends_circle_size:** Número de amigos próximos (0–15).
- **Post_frequency:** Frequência de postagens em mídias sociais (0–10).
- **Personality:** Variável alvo (Extrovert/Introvert).

**Análise Gráfica**

```python
# Gráfico 1: Distribuição da variável alvo

sns.countplot(x = 'Personality',
              data = dados,
              palette = 'pastel',
              hue='Personality')

plt.title('Distribuição da Variável Target')
plt.ylabel('Total')

plt.show()
```

```python
# Gráfico 2: Distribuição de Tempo sozinho

sns.histplot(data = dados,
             x = 'Time_spent_Alone',
             kde = True,
             color = 'skyblue',
             bins = 12)

plt.title('Horas Gastas Sozinho')
plt.xlabel('Horas')
plt.ylabel('Total')

plt.show()
```

```python
# Gráfico 3: Tempo Sozinho por Personalidade

sns.boxplot(x = 'Personality',
            y = 'Time_spent_Alone',
            data = dados,
            palette = 'pastel',
            hue = 'Personality')

plt.title('Tempo Sozinho por Tipo de Personalidade')
plt.xlabel('Personalidade')
plt.ylabel('Horas por dia')

plt.show()
```

```python
# Gráfico 4: Medo de Palco por Personalidade

sns.countplot(x = 'Stage_fear',
              hue = 'Personality',
              data = dados,
              palette = 'pastel')

plt.title('Medo de Palco por Personalidade')
plt.ylabel('Total')
plt.xlabel('Medo de Palco')

plt.show()
```

```python
# Gráfico 5: Distribuição da Frequência de Participação em Eventos Sociais

sns.histplot(data = dados,
             x = 'Social_event_attendance',
             kde = True,
             color = 'lightcoral',
             bins = 10)

plt.title('Frequência de Participação em Eventos Sociais')
plt.xlabel('Total de Eventos Sociais por Semana')
plt.ylabel('Frequência')

plt.show()
```

```python
# Gráfico 6: Eventos Sociais por Personality

plt.figure(figsize = (8,5))

sns.boxplot(x = 'Personality',
            y = 'Social_event_attendance',
            hue = 'Personality',
            data = dados,
            palette = 'pastel')

plt.title('Frequência de Participação em Eventos Sociais por Personalidade')
plt.xlabel('Personalidade')
plt.ylabel('Eventos por Semana')

plt.show()
```

```python
# Gráfico 7: Esgotamento Após Socializar por Personality

sns.countplot(x = 'Drained_after_socializing',
              hue = 'Personality',
              data = dados,
              palette = 'pastel')

plt.title('Esgotamento Após Socializar por Personalidade')
plt.ylabel('Total')
plt.xlabel('Sente-se Esgotado?')

plt.show()
```

```python
# Gráfico 8: Tamanho do Círculo Social

sns.histplot(data = dados,
             x = 'Friends_circle_size',
             kde = True,
             color = 'mediumseagreen',
             bins = 15)

plt.title('Distribuição do Número de Amigos Próximos')
plt.xlabel('Número de amigos')
plt.ylabel('Total')

plt.show()
```

```python
# Gráfico 9: Friends Circle Size por Personality

sns.boxplot(x = 'Personality',
            y = 'Friends_circle_size',
            data = dados,
            palette = 'pastel',
            hue = 'Personality')

plt.title('Tamanho do Círculo Social por Personalidade')
plt.xlabel('Personalidade')
plt.ylabel('Número de amigos')

plt.show()
```

```python
# Gráfico 10: Frequência de Postagens em Mídias Sociais

plt.figure(figsize = (8,5))

sns.histplot(data = dados,
             x = 'Post_frequency',
             kde = True,
             color = 'mediumpurple',
             bins = 10)

plt.title('Distribuição da Frequência de Postagens em Mídias Sociais')
plt.xlabel('Postagens por semana')
plt.ylabel('Frequência')

plt.show()
```

```python
# Gráfico 11: Post Frequency por Personality

sns.boxplot(x = 'Personality',
            y = 'Post_frequency',
            data = dados,
            palette = 'pastel',
            hue = 'Personality')

plt.title('Frequência de Postagens por Personalidade')
plt.xlabel('Personalidade')
plt.ylabel('Postagens por semana')

plt.show()
```

```python
# Gráfico 12: Heatmap de Correlação

plt.figure(figsize = (10,8))

corr = dados[['Time_spent_Alone', 'Social_event_attendance', 'Going_outside',
              'Friends_circle_size', 'Post_frequency']].corr()

sns.heatmap(corr,
            annot = True,
            cmap = 'coolwarm',
            fmt = ".2f",
            linewidths = 0.5)

plt.title('Matriz de Correlação')

plt.show()
```
