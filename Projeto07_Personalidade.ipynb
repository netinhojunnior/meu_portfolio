**Informações sobre as Variáveis**

- **Time_spent_Alone:** Horas gastas sozinho diariamente (0–11).
- **Stage_fear:** Presença de medo do palco (Yes/No).
- **Social_event_attendance:** Frequency of social events (0–10).
- **Going_outside:** Frequência de saídas de casa (0–7).
- **Drained_after_socializing:** Sentimento de esgotamento após socializar (Yes/No).
- **Friends_circle_size:** Número de amigos próximos (0–15).
- **Post_frequency:** Frequência de postagens em mídias sociais (0–10).
- **Personality:** Variável alvo (Extrovert/Introvert).

**Análise Gráfica**

```python
# Gráfico 1: Distribuição da variável alvo

sns.countplot(x = 'Personality',
              data = dados,
              palette = 'pastel',
              hue='Personality')

plt.title('Distribuição da Variável Target')
plt.ylabel('Total')

plt.show()
```

```python
import plotly.express as px

# Contagem da variável 'Personality'
fig = px.histogram(dados,
                   x='Personality',
                   color='Personality',
                   color_discrete_sequence=px.colors.qualitative.Pastel)

# Personalizando o layout
fig.update_layout(
    title='Distribuição da Variável Target',
    xaxis_title='Personality',
    yaxis_title='Total',
    showlegend=False
)

fig.show()
```
