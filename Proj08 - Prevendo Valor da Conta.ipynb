### **1. Introdução**

<p align="justify">
O projeto proposto tem como objetivo aplicar técnicas de machine learning, especificamente a regressão linear, para prever o valor da variável "total_bill" em um cenário de restaurante. A previsão do valor da conta total é de extrema relevância para restaurantes e profissionais da indústria de serviços, uma vez que pode proporcionar insights úteis para a gestão de negócios, estratégias de precificação e atendimento ao cliente. ### **3. Pré-Processamento de Dados**

```python
# Transformando dados categóricos em númericos

## Sex
dados['sex'] = dados['sex'].map({'Female': 0, 'Male': 1})

## Smoker
dados['smoker'] = dados['smoker'].map({'No': 0, 'Yes': 1})

## Day
dados['day'] = dados['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})

## Time
dados['time'] = dados['time'].map({'Lunch': 0, 'Dinner': 1})
```

```python
# Visualizando resultado do processamento

dados.head()
```

```python
# Separando os dados em previsores(X) e classe(Y)

previsores = np.array(dados[["tip", "sex", "smoker",
                             "day", "time", "size"]])

target = np.array(dados['total_bill'])
```

```python
# Dividindo os dados em treino(70%) e teste(30%)

X_treino, X_teste, y_treino, y_teste = train_test_split(previsores,
                                                        target,
                                                        test_size = 0.3,
                                                        random_state = 0)
```

```python
# Distribuição dos dados

print('Dados de Treino')
print(X_treino.shape)
print(y_treino.shape)
print('-----------------')
print('Dados de Teste')
print(X_teste.shape)
print(y_teste.shape)
```

### **4. Construindo Modelo**

```python
# Instanciando o modelo

modelo = LinearRegression()
```

```python
# Treinando o modelo

modelo.fit(X_treino, y_treino)
```

```python
# Realizando previsões

y_predito = modelo.predict(X_teste)
```

### **5. Avaliando e Testando Modelo**

```python
# RMSE

rmse = np.sqrt(mean_squared_error(y_teste, y_predito))
rmse
```

```python
# R²

r2 = r2_score(y_teste, y_predito)
r2
```

```python
# Aplicando o modelo a novos dados

## Novo conjunto de dados
new_data = np.array([[4.50, 1, 0, 0, 1, 4]])

## Prevendo valor da conta
modelo.predict(new_data)
```
