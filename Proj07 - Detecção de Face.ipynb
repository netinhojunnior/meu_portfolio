```python
# Carregando bibliotecas

import cv2
from google.colab.patches import cv2_imshow
```

```python
# Carregando imagem

imagem = cv2.imread('/content/drive/MyDrive/Portfólio/Dados/Pessoas.jpg')
```

```python
# Carregando o classificador de objetos pré-treinado

classicador_face = cv2.CascadeClassifier('/content/drive/MyDrive/Portfólio/Dados/haarcascade_frontalface_default.xml')
```

## **2. Pré-Processamento e Aplicação**

```python
# Convertendo a imagem para escala de cinza

gray_scale = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
# Detectando rostos na imagem

faces = classicador_face.detectMultiScale(gray_scale,
                                          scaleFactor = 1.1,
                                          minNeighbors = 5,
                                          minSize = (30, 30))
```

```python
# Desenhado retângulos ao redos dos rostos detectados

for (x, y, w, h) in faces:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
# Exibindo imagem com os retângulos desenhados

cv2_imshow(imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Ajustando os parâmetros de detecção

## Antes: minNeighbors = 5
## Agora: minNeighbors = 3

## Detectando rostos na imagem

faces = classicador_face.detectMultiScale(gray_scale,
                                          scaleFactor = 1.1,
                                          minNeighbors = 3,
                                          minSize = (30, 30))
```

```python
# Desenhando retângulos ao redor dos rostos detectados

for (x, y, w, h) in faces:
    cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
# Exibindo imagem com os retângulos desenhados

cv2_imshow(imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
